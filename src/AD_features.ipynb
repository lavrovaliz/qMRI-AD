# Exploratory analysis

```python
def run_pipeline(fs):
    
    # visualization of feature values distribution in classes (in .html report)
    fs.plot_distribution(fs._feature_column)

    # visualization of feature mutual (Spearman) correlation coefficient matrix (in .html report)
    fs.plot_correlation_matrix(fs._feature_column)

    # visualization of Mann-Whitney Bonferroni corrected p-values for binary classes test (in .html report)
    fs.plot_MW_p(fs._feature_column)

    # visualization of univariate ROC-curves (in .html report)
    fs.plot_univariate_roc(fs._feature_column)

    # calculation of basic statistics for each feature (in .xlsx):
    # number of NaN, mean, std, min, max; if applicable: MW-p, univariate ROC AUC, volume correlation
    fs.calculate_basic_stats(volume_feature='original_shape_VoxelVolume')

    # checking the excel table
    #print('Basic statistics for each feature')
    #pd.read_excel('../data/features/features_basic_stats.xlsx')

    # volume analysis
    fs.volume_analysis(volume_feature='original_shape_VoxelVolume')
    
    return None
```

```python
from statsmodels.stats import multitest
import scipy as sp
```

```python
ds_AD = pd.read_csv("Z:/Lisa/AD_raw/derivatives/hmri/MTsat_WM.csv")
ds_MS = pd.read_csv("Z:/Lisa/IV/results/features/features_CRC_MT_WM.csv")

features = ds_AD.columns[1:-1]

p_MW = []

for feature in features:
    x1 = ds_AD[feature]
    x0 = ds_MS[feature]
    s, p = sp.stats.mannwhitneyu(x1, x0, alternative='two-sided')
    p_MW.append(p)
        
reject, p_MW_corr, a_S_corr, a_B_corr = multitest.multipletests(p_MW, alpha=0.05, method='fdr_bh')

counter = 0

for i in range (0, len(p_MW_corr)):
    #if p_MW_corr[i] <= 0.05:
    #    print (features[i], p_MW_corr[i])
    #    counter += 1
    print (p_MW_corr[i])
        
#print (counter)
```

```python
def calculate_dif_mean(path_1, path_2):
    
    df_1 = pd.read_csv(path_1)
    df_2 = pd.read_csv(path_2)

    df_diff = df_1[df_1.columns[1:-1]]-df_2[df_2.columns[1:-1]]
    df_diff['sub'] = df_1['sub']
    df_diff['outcome'] = df_1['outcome']

    df_mean = (df_1[df_1.columns[1:-1]]+df_2[df_2.columns[1:-1]])/2
    df_mean['sub'] = df_1['sub']
    df_mean['outcome'] = df_1['outcome']
    
    df_diff.to_csv(path_1[:-5] + 'diff.csv')
    df_mean.to_csv(path_1[:-5] + 'mean.csv')
    
    return None
```

```python
sub_names = list(ds_age['sub'])
```

```python
def ftr_age_corr(filename_AD, filename_MS):

    ds_ftrs_AD = pd.read_csv(filename_AD)
    ds_ftrs_MS = pd.read_csv(filename_MS)

    features = ds_ftrs_AD.columns[1:-1]
    
    ds_ftrs = ds_ftrs_AD.append(ds_ftrs_MS, ignore_index = True)
    ds_ftrs.drop(ds_ftrs.loc[~ds_ftrs['sub'].isin(sub_names)].index, inplace = True)
    ds_ftrs = pd.merge(ds_ftrs, ds_age, on="sub")
    
    for feature in features:
        r, p = stats.spearmanr(ds_ftrs[feature], ds_ftrs['age'])
        if (r<0.05)&(p>0.75):
            print (feature, r, p)
```
