```python
import json
import os
import numpy as np
import pandas as pd
```

```python
scanners = []
root = "D:/CRC AD/"
for path, subdirs, files in os.walk(root):
    for name in files:
        if name[-4:] == 'json':
            try:
                with open(os.path.join(path, name)) as f:
                    data = json.load(f)
                scanner = data['acqpar'][0]['ManufacturerModelName']
                scanners.append(scanner)
            except:
                print (f)
```

```python
ages = []
root = "Z:/Lisa/AD/"
for path, subdirs, files in os.walk(root):
    for name in files:
        if name[-4:] == 'json':
            try:
                with open(os.path.join(path, name)) as f:
                    data = json.load(f)
                age = data['acqpar'][0]['PatientAge']
                pat_id = data['acqpar'][0]['PatientID']
                ages.append({'sub': pat_id, 'age': age})
            except:
                print (f)
```

```python
df_ages = pd.DataFrame(ages)
```

```python
df_ages.drop_duplicates(inplace = True)
```

```python
df_ages.to_csv('age.csv')
```
